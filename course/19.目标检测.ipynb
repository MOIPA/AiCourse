```python
def box_corner_to_center(boxes):
    """"""
    x1, y1, x2, y2 = boxes[:, 0], boxes[:, 1], boxes[:, 2], boxes[:, 3]
    cx = (x1 + x2) / 2
    cy = (y1 + y2) / 2
    w = x2 - x1
    h = y2 - y1
    boxes = torch.stack((cx, cy, w, h), axis=-1)
    return boxes

def box_center_to_corner(boxes):
    """"""
    cx, cy, w, h = boxes[:, 0], boxes[:, 1], boxes[:, 2], boxes[:, 3]
    x1 = cx - 0.5 * w
    y1 = cy - 0.5 * h
    x2 = cx + 0.5 * w
    y2 = cy + 0.5 * h
    boxes = torch.stack((x1, y1, x2, y2), axis=-1)
    return boxes
```

```python
%matplotlib inline
import os
import pandas as pd
import torch
import torchvision
from d2l import torch as d2l

d2l.DATA_HUB['banana-detection'] = (
    d2l.DATA_URL + 'banana-detection.zip',
    '5de26c8fce5ccdea9f91267273464dc968d20d72')
```

```python
def read_data_bananas(is_train=True):
    """"""
    data_dir = d2l.download_extract('banana-detection')
    csv_fname = os.path.join(data_dir, 'bananas_train' if is_train
                             else 'bananas_val', 'label.csv')
    csv_data = pd.read_csv(csv_fname)
    csv_data = csv_data.set_index('img_name'  )#set_index
    images, targets = [], []
    for img_name, target in csv_data.iterrows():
        images.append(torchvision.io.read_image(
            os.path.join(data_dir, 'bananas_train' if is_train else
                         'bananas_val', 'images', f'{img_name}')))
        # targetxyxy
        # 0
        targets.append(list(target))
        # unsqueeze11   
    return images, torch.tensor(targets).unsqueeze(1) / 256  # /256 256256
```

```python
class BananasDataset(torch.utils.data.Dataset):
    """"""
    def __init__(self, is_train):
        self.features, self.labels = read_data_bananas(is_train)
        print(f'features shape: {self.labels[0][0]}')
        print('read ' + str(len(self.features)) + (f' training examples' if
              is_train else f' validation examples'))

    def __getitem__(self, idx):
        return (self.features[idx].float(), self.labels[idx])

    def __len__(self):
        return len(self.features)
```

```python
def load_data_bananas(batch_size):
    """"""
    train_iter = torch.utils.data.DataLoader(BananasDataset(is_train=True),
                                             batch_size, shuffle=True)
    val_iter = torch.utils.data.DataLoader(BananasDataset(is_train=False),
                                           batch_size)
    return train_iter, val_iter
```

```python
# aspect ratio 
# data  sizes0-1listratios0-1list
def multibox_prior(data, sizes, ratios):
    """"""
    in_height, in_width = data.shape[-2:] #  
    device, num_sizes, num_ratios = data.device, len(sizes), len(ratios)
    boxes_per_pixel = (num_sizes + num_ratios - 1)  # 
    # print(f'boxed per pixel: {boxes_per_pixel}')
    size_tensor = torch.tensor(sizes, device=device)   # tensor
    ratio_tensor = torch.tensor(ratios, device=device)

    # 
    # 110.5
    offset_h, offset_w = 0.5, 0.5
    steps_h = 1.0 / in_height  # y
    steps_w = 1.0 / in_width  # x

    # 
    # 
    center_h = (torch.arange(in_height, device=device) + offset_h) * steps_h
    center_w = (torch.arange(in_width, device=device) + offset_w) * steps_w
    # print(f'center_h: {center_h.shape}')
    # print(f'center_w: {center_w.shape}')
    shift_y, shift_x = torch.meshgrid(center_h, center_w, indexing='ij')      # meshgridxxyyshapexxyy
    # print(f'shift_y: {shift_y.shape}')
    # print(f'shift_x: {shift_x.shape}')
    #  shift_xy
    shift_y, shift_x = shift_y.reshape(-1), shift_x.reshape(-1)           # 
    # print(f'shift_y: {shift_y[10000:10005]}')
    # print(f'shift_x: {shift_x[10000:10005]}')
    # boxes_per_pixel
    # (xmin,xmax,ymin,ymax)
    # whs*r1s1r boxes_per_pixelwboxes_per_pixel
    # print(f'size_tensor: {size_tensor.shape}')
    # print(f'ratio_tensor: {ratio_tensor.shape}')
    w = torch.cat((size_tensor * torch.sqrt(ratio_tensor[0]),
                   sizes[0] * torch.sqrt(ratio_tensor[1:])))\
                   * in_height / in_width  #   in_widthw
    h = torch.cat((size_tensor / torch.sqrt(ratio_tensor[0]),
                   sizes[0] / torch.sqrt(ratio_tensor[1:]))) # 
    # print(f'w: {w}')
    # print(f'h: {h.shape}')
    # 2
    # stack -w,-h,w,hboxes_per_pixel
    # rpeat in_height * in_width  boxes_per_pixel
    anchor_manipulations = torch.stack((-w, -h, w, h)).T.repeat(in_height * in_width, 1) / 2 #  0 in_height * in_width
    # print(f'in_height * in_width: {in_height * in_width}')
    # print(f'torch.stack((-w, -h, w, h)): {torch.stack((-w, -h, w, h)).T}')
    # print(f'anchor_manipulations shape: {anchor_manipulations.shape}')
    # print(f'anchor_manipulations: {anchor_manipulations[:10]}')


    # boxes_per_pixel
    # boxes_per_pixel
    # x,y,x,y stackxyxy boxes_per_pixel=5 55
    # out_grid stack5561*728=408408stackxy408408out_grid 5*408408 = 2042040
    out_grid = torch.stack([shift_x, shift_y, shift_x, shift_y],
                dim=1).repeat_interleave(boxes_per_pixel, dim=0)
    # print(f'torch.stack([shift_x, shift_y, shift_x, shift_y],dim=1) shape : {torch.stack([shift_x, shift_y, shift_x, shift_y],dim=1).shape}')
    # print(f'torch.stack([shift_x, shift_y, shift_x, shift_y],dim=1): {torch.stack([shift_x, shift_y, shift_x, shift_y],dim=1)[20000:20005]}')
    # print(f'out_grid: {out_grid.shape}')
    # print(f'out_grid: {out_grid[100000:100025]}')
    # x,y,x,y -w,-h,w,h 
    output = out_grid + anchor_manipulations
    # print(f'output: {output.shape}')
    # print(f'output: {output[100000:100025]}')
    # 0 2042040,4  1,2042040,4 1demo1
    # print(f'output.unsqueeze(0){output.unsqueeze(0).shape}')
    return output.unsqueeze(0)
```

```python
#
def show_bboxes(axes, bboxes, labels=None, colors=None):
    """"""
    def _make_list(obj, default_values=None):
        if obj is None:
            obj = default_values
        elif not isinstance(obj, (list, tuple)):
            obj = [obj]
        return obj

    labels = _make_list(labels)
    colors = _make_list(colors, ['b', 'g', 'r', 'm', 'c'])
    for i, bbox in enumerate(bboxes):
        color = colors[i % len(colors)]
        rect = d2l.bbox_to_rect(bbox.detach().numpy(), color)
        axes.add_patch(rect)
        if labels and len(labels) > i:
            text_color = 'k' if color == 'w' else 'w'
            axes.text(rect.xy[0], rect.xy[1], labels[i],
                      va='center', ha='center', fontsize=9, color=text_color,
                      bbox=dict(facecolor=color, lw=0))
```

```python
# 
# box_iou
#  n,m nboxes1mboxes2
def box_iou(boxes1, boxes2):
    """"""
    # box_area = 
    box_area = lambda boxes: ((boxes[:, 2] - boxes[:, 0]) *  # x2 -x1 = 
                              (boxes[:, 3] - boxes[:, 1])) # y2-y1 = 
    # boxes1,boxes2,areas1,areas2:
    # boxes1(boxes1,4),
    # boxes2(boxes2,4),
    # areas1(boxes1,),
    # areas2(boxes2,)
    areas1 = box_area(boxes1)
    areas2 = box_area(boxes2)
    print(f'boxes1: {boxes1.shape}')
    print(f'boxes2: {boxes2.shape}')
    # inter_upperlefts,inter_lowerrights,inters:
    # (boxes1,boxes2,2)
    # torch.maxboxes1 5,1,2 boxes2 3,2 boxes125,3,2 max 5,3,2
    inter_upperlefts = torch.max(boxes1[:, None, :2], boxes2[:, :2])    # None x1,y1
    inter_lowerrights = torch.min(boxes1[:, None, 2:], boxes2[:, 2:])   #  x2,y2
    inters = (inter_lowerrights - inter_upperlefts).clamp(min=0)       # 
    print(f'inter_upperlefts: {inter_upperlefts.shape}')
    print(f'inter_lowerrights: {inter_lowerrights.shape}')
    # inter_areasandunion_areas:(boxes1,boxes2)
    inter_areas = inters[:, :, 0] * inters[:, :, 1]           # 
    union_areas = areas1[:, None] + areas2 - inter_areas         #  = 
    return inter_areas / union_areas                     # / 
```

```python
# ipad
# ground_truth 
# anchors 
def assign_anchor_to_bbox(ground_truth, anchors, device, iou_threshold=0.5):
    """"""
    num_anchors, num_gt_boxes = anchors.shape[0], ground_truth.shape[0]
    # ijx_ijijIoU
    jaccard = box_iou(anchors, ground_truth)
    # 
    # -1
    anchors_bbox_map = torch.full((num_anchors,), -1, dtype=torch.long,
                                  device=device)
    # 
    # max_ious N*1indices
    max_ious, indices = torch.max(jaccard, dim=1)
    # 0
    #  anc_i
    anc_i = torch.nonzero(max_ious >= iou_threshold).reshape(-1)  # 0
    box_j = indices[max_ious >= iou_threshold]
    anchors_bbox_map[anc_i] = box_j
    col_discard = torch.full((num_anchors,), -1)
    row_discard = torch.full((num_gt_boxes,), -1)

    # 
    # 
    for _ in range(num_gt_boxes):
        max_idx = torch.argmax(jaccard)  # 
        box_idx = (max_idx % num_gt_boxes).long()   # 
        anc_idx = (max_idx / num_gt_boxes).long()   # 
        anchors_bbox_map[anc_idx] = box_idx       # 
        jaccard[:, box_idx] = col_discard   # -1
        jaccard[anc_idx, :] = row_discard   # -1
    return anchors_bbox_map
```

```python
#   
def offset_boxes(anchors, assigned_bb, eps=1e-6):
    """"""
    c_anc = d2l.box_corner_to_center(anchors) # 
    c_assigned_bb = d2l.box_corner_to_center(assigned_bb)
    # x,y /  w,h  (x2-x1,y2-y1)/(w,h)
    offset_xy = 10 * (c_assigned_bb[:, :2] - c_anc[:, :2]) / c_anc[:, 2:]
    offset_wh = 5 * torch.log(eps + c_assigned_bb[:, 2:] / c_anc[:, 2:])
    offset = torch.cat([offset_xy, offset_wh], axis=1)
    return offset
```

```python
# 
# 0
# xy
# 
# 0mask01
# *mask
#  mask
def multibox_target(anchors, labels):
    print(f'anchors shape{anchors.shape}')
    print(f'labels shape{labels.shape}')
    """"""
    batch_size, anchors = labels.shape[0], anchors.squeeze(0) # 1
    batch_offset, batch_mask, batch_class_labels = [], [], []
    device, num_anchors = anchors.device, anchors.shape[0]
    for i in range(batch_size):
        label = labels[i, :, :] #   n,5 n 5
        print(f'label {label}')
        anchors_bbox_map = assign_anchor_to_bbox(
            label[:, 1:], anchors, device)
        print(f'anchors_bbox_map {anchors_bbox_map}')
        # 5540
        # 
        bbox_mask = ((anchors_bbox_map >= 0).float().unsqueeze(-1)).repeat(1, 4) # 
        print(f'bbox_mask {bbox_mask}')
        # 
        class_labels = torch.zeros(num_anchors, dtype=torch.long,
                                   device=device)
        assigned_bb = torch.zeros((num_anchors, 4), dtype=torch.float32,
                                  device=device)
        # 
        # 
        indices_true = torch.nonzero(anchors_bbox_map >= 0)
        print(f'indices_true {indices_true}')
        # id
        bb_idx = anchors_bbox_map[indices_true]
        print(f'bb_idx {bb_idx}')
        # labelclass_labels0
        class_labels[indices_true] = label[bb_idx, 0].long() + 1
        print(f'class_labels {class_labels}')
        # 0
        assigned_bb[indices_true] = label[bb_idx, 1:]
        print(f'assigned_bb {assigned_bb}')

        #   0
        #  bbox_mask 0
        offset = offset_boxes(anchors, assigned_bb) * bbox_mask
        print(f'offset {offset}')

        batch_offset.append(offset.reshape(-1))
        batch_mask.append(bbox_mask.reshape(-1))
        batch_class_labels.append(class_labels)
        print(f'batch_offset {batch_offset}')
    bbox_offset = torch.stack(batch_offset)
    bbox_mask = torch.stack(batch_mask)
    class_labels = torch.stack(batch_class_labels)
    # maskoffset*maskmask00
    #   bbox_mask 0
    return (bbox_offset, bbox_mask, class_labels)
```

```python
def offset_inverse(anchors, offset_preds):
    """"""
    anc = d2l.box_corner_to_center(anchors)
    pred_bbox_xy = (offset_preds[:, :2] * anc[:, 2:] / 10) + anc[:, :2]
    pred_bbox_wh = torch.exp(offset_preds[:, 2:] / 5) * anc[:, 2:]
    pred_bbox = torch.cat((pred_bbox_xy, pred_bbox_wh), axis=1)
    predicted_bbox = d2l.box_center_to_corner(pred_bbox)
    return predicted_bbox
```

```python
# scores
# boxes n,4 n
# iou_threshold 
#  [0,3]
def nms(boxes, scores, iou_threshold):
    """"""
    B = torch.argsort(scores, dim=-1, descending=True)
    keep = []  # 
    while B.numel() > 0:
        i = B[0]
        keep.append(i)
        if B.numel() == 1: break
        iou = box_iou(boxes[i, :].reshape(-1, 4),
                      boxes[B[1:], :].reshape(-1, 4)).reshape(-1)
        # 
        inds = torch.nonzero(iou <= iou_threshold).reshape(-1) # -1
        B = B[inds + 1]
    return torch.tensor(keep, device=boxes.device)
```

```python
# 
# cls_probs 134 4
# anchors 144 4
# offset_preds  116   4*4 = 16 
# nms_threshold 0.5
# pos_threshold 

#  6label -1)
def multibox_detection(cls_probs, offset_preds, anchors, nms_threshold=0.5,
                       pos_threshold=0.009999999):
    """"""
    device, batch_size = cls_probs.device, cls_probs.shape[0]
    anchors = anchors.squeeze(0) # 
    num_classes, num_anchors = cls_probs.shape[1], cls_probs.shape[2]
    out = []
    for i in range(batch_size):
        cls_prob, offset_pred = cls_probs[i], offset_preds[i].reshape(-1, 4)
        #   cls_prob
        # 00 class_id
        conf, class_id = torch.max(cls_prob[1:], 0)
        # 
        predicted_bb = offset_inverse(anchors, offset_pred)
        # nms
        keep = nms(predicted_bb, conf, nms_threshold)

        # non_keep
        all_idx = torch.arange(num_anchors, dtype=torch.long, device=device)
        # [0,3]03 0123  0,3,0,1,2,3
        # 0123
        combined = torch.cat((keep, all_idx))
        # uniques0123counts21121
        uniques, counts = combined.unique(return_counts=True)
        # 1 12
        non_keep = uniques[counts == 1]
        all_id_sorted = torch.cat((keep, non_keep))  # 0312   0312
        class_id[non_keep] = -1                     # -1
        class_id = class_id[all_id_sorted]          #   
        conf, predicted_bb = conf[all_id_sorted], predicted_bb[all_id_sorted]  # 0312 
        # pos_threshold
        below_min_idx = (conf < pos_threshold)
        # -1
        class_id[below_min_idx] = -1
        conf[below_min_idx] = 1 - conf[below_min_idx]
        # unsqueeze 
        # pred_info
        pred_info = torch.cat((class_id.unsqueeze(1),
                               conf.unsqueeze(1),
                               predicted_bb), dim=1)
        out.append(pred_info)
    return torch.stack(out)
```
